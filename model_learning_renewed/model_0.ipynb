```python
import pandas as pd
import numpy as np
import torch
from torch import nn
```

```python
train_data = torch.tensor([[1,0],[1,1],[0,0],[0,1]], dtype=torch.float) # in_features = 2 
train_labels = torch.tensor([[0],[1],[1],[0]], dtype=torch.float) # out_features = 1
```

```python
def display_frame(data,*args):
    temp = pd.DataFrame(data)
    for temp_1 in args:
        temp.insert(temp.shape[1], temp.shape[1], temp_1)
    print(temp)
    
# train_df = pd.DataFrame(train_data.numpy())
# train_df.insert(train_df.shape[1],"labels", train_labels.numpy())
# train_df

display_frame(train_data,train_labels)
```

```python
class Model(nn.Module):
    def __init__(self,in_f,out_f):
        super(Model,self).__init__()
        
        self.linear_1 = nn.Linear(in_f,2)
        self.relu = nn.ReLU()
        self.linear_2 = nn.Linear(2,4)
        self.linear_3 = nn.Linear(4,out_f)

    def forward(self,x):
        print("Model: ")
        print("Shape of x: ",x.shape)
        x = self.linear_1(x)
        print("Shape of x after linear_1: ", x.shape)
        x = self.relu(x)
        print("Shape of x after relu: ", x.shape)
        x = self.linear_2(x)
        print("Shape of x after linear_2: ", x.shape)
        x = self.relu(x)
        print("Shape of x after relu: ", x.shape)
        x = self.linear_3(x)
        print("Shape of x after linear_3: ", x.shape)
        x = self.relu(x)
        print("Shape of x after relu: ", x.shape)
        return x
```

```python
torch.manual_seed(42)
model = Model(2,1)
```

```python
list(model.parameters())
```

```python
list(model.parameters())
```

```python
print(model.linear_1.weight)
print(model.linear_1.bias)
```

```python
from torchsummary import summary
summary(model, input_size=(4,2))
```

```python
loss_fn = nn.L1Loss()
optimizer = torch.optim.SGD(params=model.parameters(), lr=0.01)
```

```python

y_pred = model(train_data)
y_pred
```

```python
epochs = 100

for i in range(epochs):
    model.train()

    y_pred = model(train_data)

    loss = loss_fn(y_pred,train_labels)

    optimizer.zero_grad()
    
    loss.backward()

    optimizer.step()

    model.eval()
    
```

```python

y_pred = model(train_data)
y_pred, train_labels
```
