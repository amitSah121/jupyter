```python
def find_classes(dir: str) -> Tuple[List[str], Dict[str,int]] :
    classes = sorted([entry.name for entry in list(os.scandir(dir)) if entry.is_dir()])

    if not classes:
        print("Couldn't find any classes in target directory")
        return None
    class_idx = {class_name: i for i,class_name in enumerate(classes)}
    return classes, class_idx
```

```python
class ImageFolderCustom(Dataset):
    def __init__(self, target_dir: str, transform: None):
        self.classes = []
        self.class_idx = {}
        self.paths = []
        self.p,self.q = find_classes(target_dir)
        for path in image_dirs:
            image_paths = list(path.glob("*.jpg"))
            for image_path in image_paths:
                self.paths.append(image_path)
                self.classes.append(image_path.name)
                self.class_idx[image_path.name] = self.q[path.name]
        # self.classes, self.class_idx = find_classes(target_dir)
        self.transform = transform

    def loadImage(self, index):
        img = self.paths[index]
        return Image.open(img)

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, index) -> Tuple[torch.Tensor, int]:
        img = self.loadImage(index)
        class_name = self.paths[index].name
        class_idx = self.class_idx[class_name]

        if self.transform:
            return self.transform(img), class_idx
        else:
            return img, class_idx
```

```python
class VGG16(nn.Module):
    def __init__(self, num_classes=10):
        super(VGG16, self).__init__()
        self.layer1 = nn.Sequential(  # 64,3,64,64 --> 64,(64),64,64
            nn.Conv2d(in_channels = 3, out_channels = 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU())
        print("layer1: ",self.layer1.parameters)
        
        self.layer2 = nn.Sequential( # 64,(64),64,64 --> 64,(64),64,64
            nn.Conv2d(64, 64, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(), 
            nn.MaxPool2d(kernel_size = 2, stride = 2)) # 64,(64),64,64 --> 64,(64),32,32
        print("layer2: ",self.layer2.parameters)
        
        self.layer3 = nn.Sequential( #64,(64),32,32 --> 64,(128),32,32
            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU())
        print("layer3: ",self.layer3.parameters)
        
        self.layer4 = nn.Sequential( # #64,(128),32,32 --> 64,(128),32,32
            nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2)) # #64,(128),32,32 --> 64,(128),16,16
        print("layer4: ",self.layer4.parameters)
        
        self.layer5 = nn.Sequential( # 64,(128),16,16 --> 64,(256),16,16
            nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU())
        print("layer5: ",self.layer5.parameters)
        
        self.layer6 = nn.Sequential( # 64,(256),16,16 --> 64,(256),16,16
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU())
        print("layer6: ",self.layer6.parameters)
        
        self.layer7 = nn.Sequential( # 64,(256),16,16 --> 64,(256),16,16
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2)) # 64,(256),16,16 --> 64,(256),8,8
        print("layer7: ",self.layer7.parameters)
        
        self.layer8 = nn.Sequential( # 64,(256),8,8 --> 64,(512),8,8
            nn.Conv2d(256, 512, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU())
        print("layer8: ",self.layer8.parameters)
        
        self.layer9 = nn.Sequential( # 64,(512),8,8 --> 64,(512),8,8
            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU())
        print("layer9: ",self.layer9.parameters)
        
        self.layer10 = nn.Sequential( # 64,(512),8,8 --> 64,(512),8,8
            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2)) # 64,(256),8,8 --> 64,(512),4,4
        print("layer10: ",self.layer10.parameters)
        
        self.layer11 = nn.Sequential( # 64,(512),4,4 --> 64,(512),4,4
            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU())
        print("layer11: ",self.layer11.parameters)
        
        self.layer12 = nn.Sequential( # 64,(512),4,4 --> 64,(512),4,4
            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU())
        print("layer12: ",self.layer12.parameters)
        
        self.layer13 = nn.Sequential( # 64,(512),4,4 --> 64,(512),4,4
            nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size = 2, stride = 2)) # 64,(512),4,4 --> 64,(512),2,2
        print("layer13: ",self.layer1.parameters)
        
        self.fc = nn.Sequential( # 512*2*2 = 2048
            nn.Dropout(0.5),
            nn.Linear(512*4*4, 2048),
            nn.ReLU())
        print("fc: ",self.fc.parameters)
        
        self.fc1 = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(2048, 4096),
            nn.ReLU())
        print("fc1: ",self.fc1.parameters)
        
        self.fc2= nn.Sequential(
            nn.Linear(4096, num_classes))
        print("fc2: ",self.fc2.parameters)
        
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        out = self.layer5(out)
        out = self.layer6(out)
        out = self.layer7(out)
        out = self.layer8(out)
        out = self.layer9(out)
        out = self.layer10(out)
        out = self.layer11(out)
        out = self.layer12(out)
        out = self.layer13(out)
        out = out.reshape(out.size(0), -1)
        out = self.fc(out)
        out = self.fc1(out)
        out = self.fc2(out)
        return out
#         return self.fc2(self.fc1(self.fc(self.layer13(self.layer12(self.layer11(self.layer10(self.layer9(self.layer8(self.layer7(self.layer6(self.layer5(self.layer4(self.layer3(self.layer2(self.layer1(x))))))))))))).reshape(0,-1))))
```
