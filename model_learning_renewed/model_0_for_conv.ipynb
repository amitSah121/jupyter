```python
import pandas as pd
import numpy as np
import torch
from torch import nn
```

```python
train_data = torch.tensor([[[1],[0]],[[1],[1]],[[0],[0]],[[0],[1]]], dtype=torch.float) # in_features = 2 
train_labels = torch.tensor([[0],[1],[1],[0]], dtype=torch.float) # out_features = 1
```

```python
train_data.shape, train_labels.shape
```

```python
def display_frame(data,*args):
    temp = pd.DataFrame(data)
    for temp_1 in args:
        temp.insert(temp.shape[1], temp.shape[1], temp_1)
    print(temp)
    
# train_df = pd.DataFrame(train_data.numpy())
# train_df.insert(train_df.shape[1],"labels", train_labels.numpy())
# train_df

display_frame(train_data,train_labels)
```

```python
class Model(nn.Module):
    def __init__(self,in_f,out_f):
        super(Model,self).__init__()
        
        self.conv1 = nn.Conv1d(in_f,10, kernel_size=1,stride=1)
        self.conv2 = nn.Conv1d(10,out_f, kernel_size=1,stride=1)
        self.relu = nn.ReLU()

    def forward(self,x):
        print("Model: ")
        print("Shape of x: ",x.shape)
        x = self.conv1(x)
        print("Shape of x after conv1: ", x.shape)
        x = self.conv2(x)
        print("Shape of x after conv2: ", x.shape)
        x = self.relu(x)
        return x
```

```python
torch.manual_seed(45)
model = Model(2,1)
```

```python
loss_fn = nn.L1Loss()
# optimizer = torch.optim.SGD(params=model.parameters(), lr=0.01)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
# loss_fn = nn.CrossEntropyLoss()
```

```python

y_pred = model(train_data[1])
y_pred.squeeze(dim=0), train_labels[1], loss_fn(y_pred.squeeze(dim=0), train_labels[0])
```

```python
epochs = 1000

for i in range(epochs):
    # for j in range(4):
    model.train()

    y_pred = model(train_data)

    loss = loss_fn(y_pred.squeeze(dim=0),train_labels)
    print(loss)

    optimizer.zero_grad()
    
    loss.backward()

    optimizer.step()

    model.eval()
    
```
